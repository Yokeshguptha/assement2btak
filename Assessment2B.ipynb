###### Department wise codekata performence (pie chart)

```python
df3["Department"].value_counts()
```

```python
explode =[0.05,0.04,0.03]
df3.groupby(['Department']).sum().plot(kind='pie', y='CodeKata Score', autopct='%1.2f%%', explode=explode, shadow=True)
plt.title('DEPARTMENT WISE CODEKATA PERFORMANCE')
plt.legend()
```
